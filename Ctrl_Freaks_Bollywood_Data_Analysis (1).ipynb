Below code is used to load multiple CSV files from a specific folder in Google Drive into separate Pandas DataFrames. Here's what it does:

1. **Google Drive Integration**: Mounts Google Drive to access files stored in a specific folder.
2. **Folder Traversal**: Recursively searches for all `.csv` files in the specified directory and its subdirectories.
3. **DataFrame Creation**: For each `.csv` file found, dynamically creates a DataFrame using the base filename (without extension) as the variable name and loads the data into it.
4. **Error Handling**: Prints errors for any issues encountered while loading specific files.
5. **Preview**: Optionally previews one of the loaded DataFrames (e.g., `bollywood_crew`).


```python
import pandas as pd
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the path to the folder containing the CSV files
base_path = '/content/drive/My Drive/1950-2019'

# Recursively find all .csv files in the folder
csv_file_paths = []
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith('.csv'):
            csv_file_paths.append(os.path.join(root, file))

# Check if any .csv files are found
if not csv_file_paths:
    raise FileNotFoundError("No CSV files found in the specified directory and subdirectories.")

# Load each CSV into a separate DataFrame
for path in csv_file_paths:
    try:
        # Use the base filename (without extension) as the DataFrame name
        file_name = os.path.splitext(os.path.basename(path))[0]

        # Dynamically create variables for each DataFrame
        globals()[file_name] = pd.read_csv(path, encoding='utf-8')
        print(f"Successfully loaded: {file_name}")
    except Exception as e:
        print(f"Error loading {path}: {e}")

# Example: Accessing one of the DataFrames
# Print the first few rows of one DataFrame
print("\nPreview of one DataFrame:")
print(bollywood_crew.head())  # Replace with another DataFrame name if needed
```

Below code preprocesses Bollywood datasets by cleaning and structuring the data.

It standardizes column names, splits `wins_nominations` into `wins` and `nominations`, removes unwanted rows, handles missing values, and processes specific columns like `release_date`.

The cleaned data is saved as CSV files in a designated Google Drive folder for further analysis or visualization.

```python
# Define the output folder in Google Drive
output_folder = '/content/drive/My Drive/1950-2019/preprocessed'
os.makedirs(output_folder, exist_ok=True)  # Create the folder if it doesn't exist


def split_wins_nominations(df):
    """
    Splits the 'wins_nominations' column into 'wins' and 'nominations'.
    """
    if 'wins_nominations' in df.columns:
        df['wins'] = df['wins_nominations'].str.extract(r'(\d+)\s+wins')[0].fillna(0).astype(int)
        df['nominations'] = df['wins_nominations'].str.extract(r'(\d+)\s+nomination')[0].fillna(0).astype(int)
        df.drop(columns=['wins_nominations'], inplace=True)
    return df



# Process bollywood_crew
def preprocess_bollywood_crew(df):
    """
    Specific preprocessing for bollywood_crew.
    """
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)

    # Remove rows with '\n' in any column
    df = df[~df.apply(lambda x: x.astype(str).str.contains(r'\\N').any(), axis=1)]

    # Drop rows with missing values in essential columns
    essential_columns = ['imdb_id',	'directors',	'writers']
    for col in essential_columns:
        if col in df.columns:
            df = df.dropna(subset=[col])

    # Fill missing values in numeric columns
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].mean())
    df = split_wins_nominations(df)
    return df

bollywood_crew_df = preprocess_bollywood_crew(bollywood_crew)
bollywood_crew_output_path = os.path.join(output_folder, "bollywood_crew_preprocessed.csv")
bollywood_crew_df.to_csv(bollywood_crew_output_path, index=False)
print(f"Processed and saved bollywood_crew to {bollywood_crew_output_path}")

# Process bollywood_full
def preprocess_bollywood_full(df):
    """
    Specific preprocessing for bollywood_full.
    """
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)

    # Remove rows with '\n' in any column
    df = df[~df.apply(lambda x: x.astype(str).str.contains(r'\\N').any(), axis=1)]

    if 'summary' in df.columns:
        df.drop(columns=['summary'], inplace=True)

    # Drop rows with missing values in essential columns
    essential_columns = ['title', 'release_date', 'genre', 'imdb_rating']
    for col in essential_columns:
        if col in df.columns:
            df = df.dropna(subset=[col])

    # Fill missing values in numeric columns
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].mean())

    if 'release_date' in df.columns:
        # Extract location part from the original release_date column
        df['location'] = df['release_date'].str.extract(r'\((.*?)\)')[0]

    if 'release_date' in df.columns:
          df['release_date'] = df['release_date'].str.extract(r'(\d{1,2}\s+\w+\s+\d{4})')[0] # Extract date part
          df['release_date'] = pd.to_datetime(df['release_date'], format='%d %B %Y', errors='coerce').dt.strftime('%m/%d/%Y')

      # Extract location into a new column
    df = split_wins_nominations(df)

    return df

bollywood_full_df = preprocess_bollywood_full(bollywood_full)
bollywood_full_output_path = os.path.join(output_folder, "bollywood_full_preprocessed.csv")
bollywood_full_df.to_csv(bollywood_full_output_path, index=False)
print(f"Processed and saved bollywood_full to {bollywood_full_output_path}")

# Process bollywood_writers_data
def preprocess_bollywood_writers_data(df):
    """
    Specific preprocessing for bollywood_writers_data.
    """
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)

    # Remove rows with '\n' in any column
    df = df[~df.apply(lambda x: x.astype(str).str.contains(r'\\N').any(), axis=1)]

    # Drop rows with missing values in essential columns
    essential_columns = ['crew_id',	'name']
    for col in essential_columns:
        if col in df.columns:
            df = df.dropna(subset=[col])

    # Fill missing values in numeric columns
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].mean())

    df = split_wins_nominations(df)

    return df

bollywood_writers_data_df = preprocess_bollywood_writers_data(bollywood_writers_data)
bollywood_writers_output_path = os.path.join(output_folder, "bollywood_writers_data_preprocessed.csv")
bollywood_writers_data_df.to_csv(bollywood_writers_output_path, index=False)
print(f"Processed and saved bollywood_writers_data to {bollywood_writers_output_path}")

# Process bollywood_crew_data
def preprocess_bollywood_crew_data(df):
    """
    Specific preprocessing for bollywood_crew_data.
    """
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)

    # Remove rows with '\n' in any column
    df = df[~df.apply(lambda x: x.astype(str).str.contains(r'\\N').any(), axis=1)]

    # Drop rows with missing values in essential columns
    essential_columns = ['crew_id', 'name']
    for col in essential_columns:
        if col in df.columns:
            df = df.dropna(subset=[col])

    # Fill missing values in numeric columns
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].mean())
    df = split_wins_nominations(df)

    return df

bollywood_crew_data_df = preprocess_bollywood_crew_data(bollywood_crew_data)
bollywood_crew_data_output_path = os.path.join(output_folder, "bollywood_crew_data_preprocessed.csv")
bollywood_crew_data_df.to_csv(bollywood_crew_data_output_path, index=False)
print(f"Processed and saved bollywood_crew_data to {bollywood_crew_data_output_path}")

# Preview one of the preprocessed DataFrames
print("\nPreview of bollywood_crew_df:")
print(bollywood_crew_df.head())
```

Below code processes the Bollywood dataset by splitting the `genres` and `actors` columns into separate columns for easier analysis.

It fills missing values in these columns with empty strings, splits the values using the pipe (`|`) delimiter, and renames the new columns as `genre1`, `genre2`, etc., and `actor1`, `actor2`, etc.

The original `genres` and `actors` columns are then dropped, and the updated DataFrame is saved as a new CSV file for further use.

```python
### splitting the genres and actors

import pandas as pd

# Load the dataset
df = pd.read_csv("/content/drive/My Drive/1950-2019/preprocessed/bollywood_full_preprocessed.csv")

# Ensure genres and actors columns are filled
df['genres'] = df['genres'].fillna('')
df['actors'] = df['actors'].fillna('')

# Split genres into separate columns
genre_split = df['genres'].str.split('|', expand=True)
genre_split.columns = [f'genre{i+1}' for i in range(genre_split.shape[1])]

# Split actors into separate columns
actor_split = df['actors'].str.split('|', expand=True)
actor_split.columns = [f'actor{i+1}' for i in range(actor_split.shape[1])]

# Concatenate the original dataframe with the split columns
df_split = pd.concat([df, genre_split, actor_split], axis=1)

# Drop the original 'genres' and 'actors' columns
df_split = df_split.drop(columns=['genres', 'actors'])

# Save the updated DataFrame back to a CSV file
df_split.to_csv("/content/drive/My Drive/1950-2019/preprocessed/bollywood_full_preprocessed_split.csv", index=False)

df_split.head()
```

Below code prepares a dataset to analyze the timeline of genre popularity.

It first melts the genre-related columns (`genre1`, `genre2`, etc.) into a single `genre` column while retaining `imdb_id` and `year_of_release` for context.

Next, it groups the data by `year_of_release` and `genre` to calculate the count of movies for each genre per year.

Finally, the resulting dataset, which captures genre popularity trends over time, is exported as a CSV file for further analysis.

```python
#### prepare dataset for timeline of genre popularity
# Melt genre columns into a single column
genre_columns = ['genre1', 'genre2', 'genre3']  # Adjust if actual column names differ
genres_exploded = df_split.melt(id_vars=['imdb_id', 'year_of_release'],
                              value_vars=genre_columns,
                              var_name='genre_type',
                              value_name='genre').dropna()

# Group by year and genre to calculate counts
genre_timeline = genres_exploded.groupby(['year_of_release', 'genre']).size().reset_index(name='count')

# Export the timeline data
genre_timeline.to_csv("/content/drive/My Drive/1950-2019/preprocessed/genre_timeline.csv", index=False)
```

This code integrates data from two datasets: one containing movie crew details (e.g., directors) and the other containing information about actors and genres.

The datasets are merged using a common identifier (`imdb_id`) to consolidate information for each movie.

Additionally, the `directors` column is processed to ensure that multiple directors for a movie are stored as lists, enabling easier analysis of director collaborations and contributions.

```python
import pandas as pd

# Load the datasets
crew_data_path = "/content/drive/My Drive/1950-2019/preprocessed/bollywood_crew_preprocessed.csv"
actors_genres_path = "/content/drive/My Drive/1950-2019/preprocessed/bollywood_full_preprocessed_split.csv"

crew_df = pd.read_csv(crew_data_path)
actors_genres_df = pd.read_csv(actors_genres_path)

# Merge datasets on `imdb_id`
merged_df = actors_genres_df.merge(crew_df[['imdb_id', 'directors']], on='imdb_id', how='inner')

# Ensure directors are split into lists
merged_df['directors'] = merged_df['directors'].fillna('').str.split('|')
```

Below code generates a comprehensive list of actor-director collaborations for each movie.

It identifies actors from multiple columns (`actor1`, `actor2`, etc.) and directors from the `directors` column, creating all possible actor-director pairings using the Cartesian product.

Each pairing is associated with the corresponding `imdb_id` for the movie.

The resulting data is saved as a new dataset, `actor_director_collaboration.csv`, enabling further analysis of collaborations between actors and directors in Bollywood.

```python
from itertools import product

# Generate actor-director pairs
actor_columns = [f'actor{i}' for i in range(1, 16)]
actor_director_pairs = []

for _, row in merged_df.iterrows():
    actors = row[actor_columns].dropna().tolist()
    directors = row['directors']
    pairs = product(actors, directors)  # Create all combinations of actors and directors
    actor_director_pairs.extend([(row['imdb_id'], actor, director) for actor, director in pairs])

# Create DataFrame
actor_director_collab = pd.DataFrame(actor_director_pairs, columns=['imdb_id', 'actor', 'director'])

# Save the data
actor_director_collab.to_csv("/content/drive/My Drive/1950-2019/preprocessed/actor_director_collaboration.csv", index=False)

actor_director_collab.head()
```

Below code visualizes the **top 10 actor-director collaborations** using a directed graph.

Loads a dataset and ensures the collaboration count (`collaboration_count`) column is present. If not, it calculates the count of collaborations for each actor-director pair.

Extracts the top 10 actor-director pairs with the highest collaboration counts.

Constructs a directed graph using NetworkX, where nodes represent actors and directors, and edges represent collaborations, weighted by the collaboration count.

Adjusts node sizes and colors based on their degree and edge weights. Uses `spring_layout` for better node spacing.

Plots the graph with node labels, edge labels (collaboration counts), and a visually appealing layout using Matplotlib.

The graph highlights the most frequent collaborations between actors and directors, making it easier to identify key patterns and relationships.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset
frequent_collaborations = pd.read_csv('/content/drive/My Drive/1950-2019/preprocessed/merged_actor_director_data.csv')

# Ensure collaboration_count exists or calculate it if needed
if 'collaboration_count' not in frequent_collaborations.columns:
    frequent_collaborations = frequent_collaborations.groupby(['actor', 'name']).size().reset_index(name='collaboration_count')

# Sort by collaboration count and get the top 10
top_10_collaborations = frequent_collaborations.nlargest(10, 'collaboration_count')

# Create a directed graph
G = nx.DiGraph()

# Add edges to the graph (actor-director collaborations)
for _, row in top_10_collaborations.iterrows():
    G.add_edge(row['actor'], row['name'], weight=row['collaboration_count'])

# Adjust node sizes and colors
node_sizes = [1000 + G.degree(node) * 300 for node in G.nodes()]
node_colors = [G.degree(node, weight='weight') for node in G.nodes()]

# Use spring_layout for better spacing
pos = nx.spring_layout(G, seed=42, k=0.9)

# Plot the graph
plt.figure(figsize=(16, 12))
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.viridis, alpha=0.8)
nx.draw_networkx_edges(G, pos, width=2, edge_color='gray', alpha=0.7, arrowsize=20, arrowstyle='->')
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Add edge labels for collaboration count
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Title and Layout
plt.title("Top 10 Frequent Actor-Director Collaborations (Enhanced)", fontsize=18)
plt.axis('off')
plt.tight_layout()
plt.show()
```

Below code analyzes and visualizes the top 10 actor-director pairs in Bollywood based on the number of awards ("wins").

A horizontal bar chart is then created to display these top collaborations, with clear labels and an inverted y-axis for better readability.

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
bollywood_full_preprocessed_split_path = '/content/drive/My Drive/1950-2019/preprocessed/bollywood_full_preprocessed_split.csv'
merged_actor_director_path = '/content/drive/My Drive/1950-2019/preprocessed/merged_actor_director_data.csv'

# Load the datasets
bollywood_full_preprocessed_split = pd.read_csv(bollywood_full_preprocessed_split_path)
merged_actor_director = pd.read_csv(merged_actor_director_path)

# Merge the datasets on the common column 'imdb_id'
combined_data = pd.merge(merged_actor_director, bollywood_full_preprocessed_split, on='imdb_id', how='inner')

# Ensure the 'wins' column exists and process
if 'wins' in combined_data.columns:
    # Convert 'wins' column to numeric, if necessary
    combined_data['wins'] = pd.to_numeric(combined_data['wins'], errors='coerce').fillna(0)

    # Group by actor and director, and sum the 'wins'
    grouped_data = combined_data.groupby(['actor', 'name']).sum().reset_index()

    # Sort by wins and get the top 10
    top_10_wins = grouped_data.nlargest(10, 'wins')

    # Create a bar chart
    plt.figure(figsize=(12, 8))
    plt.barh(
        [f"{row['actor']} - {row['name']}" for _, row in top_10_wins.iterrows()],
        top_10_wins['wins'],
        color='skyblue',
        edgecolor='black'
    )

    # Add labels and title
    plt.title('Top 10 Actor-Director Pairs with Most Wins', fontsize=16)
    plt.xlabel('Number of Wins', fontsize=14)
    plt.ylabel('Actor-Director Pair', fontsize=14)
    plt.gca().invert_yaxis()  # Invert y-axis for better readability
    plt.tight_layout()
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("The 'wins' column is missing in the dataset.")
```

Below code identifies the **top 10 actor-director pairs** with the most wins and their **average IMDb ratings**.

It merges datasets, processes data, calculates wins and ratings, and visualizes the results using a **bubble chart** to highlight successful collaborations and their quality.

```python
import pandas as pd

# File paths
bollywood_full_preprocessed_split_path = '/content/drive/My Drive/1950-2019/preprocessed/bollywood_full_preprocessed_split.csv'
merged_actor_director_path = '/content/drive/My Drive/1950-2019/preprocessed/merged_actor_director_data.csv'

# Load the datasets
bollywood_full_preprocessed_split = pd.read_csv(bollywood_full_preprocessed_split_path)
merged_actor_director = pd.read_csv(merged_actor_director_path)

# Merge the datasets on the common column 'imdb_id'
combined_data = pd.merge(merged_actor_director, bollywood_full_preprocessed_split, on='imdb_id', how='inner')

# Ensure the 'wins' and 'imdb_rating' columns exist
if 'wins' in combined_data.columns and 'imdb_rating' in combined_data.columns:
    # Convert 'wins' and 'imdb_rating' columns to numeric
    combined_data['wins'] = pd.to_numeric(combined_data['wins'], errors='coerce').fillna(0)
    combined_data['imdb_rating'] = pd.to_numeric(combined_data['imdb_rating'], errors='coerce').fillna(0)

    # Group by actor and director, summing the 'wins' and calculating average IMDb rating
    grouped_data = combined_data.groupby(['actor', 'name']).agg(
        total_wins=('wins', 'sum'),
        avg_imdb_rating=('imdb_rating', 'mean')
    ).reset_index()

    # Sort by total_wins and get the top 10 actor-director pairs
    top_10_pairs = grouped_data.nlargest(10, 'total_wins')

    # Display the result
    print("\nTop 10 Actor-Director Pairs with Average IMDb Ratings:")
    print(top_10_pairs)

    # Save the results to a CSV file (optional)
    output_path = '/content/drive/My Drive/1950-2019/preprocessed/top_10_avg_imdb_rating.csv'
    top_10_pairs.to_csv(output_path, index=False)
    print(f"\nResults saved to: {output_path}")
else:
    print("The 'wins' or 'imdb_rating' column is missing in the dataset.")

import matplotlib.pyplot as plt

# Plot the results
if not top_10_pairs.empty:
    plt.figure(figsize=(12, 8))

    # Scatter plot
    plt.scatter(
        range(len(top_10_pairs)),
        top_10_pairs['avg_imdb_rating'],
        s=550,  # Bubble size
        color='royalblue',
        edgecolor='black',
        alpha=0.8
    )

    # Add labels
    for i, (pair, rating) in enumerate(zip(
            [f"{row['actor']} - {row['name']}" for _, row in top_10_pairs.iterrows()],
            top_10_pairs['avg_imdb_rating'])):
        plt.text(i, rating + 0.1, f"{rating:.1f}", ha='center', fontsize=10)

    # Customize the plot
    plt.xticks(
        range(len(top_10_pairs)),
        [f"{row['actor']} - {row['name']}" for _, row in top_10_pairs.iterrows()],
        rotation=45,
        ha='right',
        fontsize=10
    )
    plt.title('Average IMDb Rating for Top 10 Actor-Director Pairs', fontsize=16)
    plt.xlabel('Actor-Director Pair', fontsize=14)
    plt.ylabel('Average IMDb Rating', fontsize=14)
    plt.ylim(0, max(top_10_pairs['avg_imdb_rating']) + 1)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("No data available to plot.")

```

### Insights:
